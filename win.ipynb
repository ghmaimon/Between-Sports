```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

mlb_df=pd.read_csv("assets/mlb.csv")
nhl_df=pd.read_csv("assets/nhl.csv")
nba_df=pd.read_csv("assets/nba.csv")
nfl_df=pd.read_csv("assets/nfl.csv")
cities=pd.read_html("assets/wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
```

```python

def mlb(): 
    # YOUR CODE HERE
    global mlb_df
    global cities
    cities.replace("^[\s*—*\ *]$",np.nan,regex = True,inplace = True)
    cities.replace("\[.*\]","",regex = True,inplace = True)
    cities.replace("^$",np.nan,regex = True,inplace = True)
    cities["Metropolitan area"] = cities["Metropolitan area"].str.strip()
    cities.iloc[13,2] = np.nan
    def city_associate(row):
        if len(row["team"].split(" ")) == 2:
            row["Metropolitan area"] = row["team"].split(" ")[0]
        if re.search("New York",row["team"]) or re.search("New Jersey",row["team"]):
            row["Metropolitan area"] = "New York City"
        if re.search("Los Angeles",row["team"]) or re.search("Anaheim",row["team"]):
            row["Metropolitan area"] = "Los Angeles"
        if re.search("Toronto",row["team"]):
            row["Metropolitan area"] = "Toronto"
        if re.search("Detroit",row["team"]):
            row["Metropolitan area"] = "Detroit"
        if re.search("Tampa",row["team"]):
            row["Metropolitan area"] = "Tampa Bay Area"
        if re.search("Columbus",row["team"]):
            row["Metropolitan area"] = "Columbus"
        if re.search("San Jose",row["team"]) or re.search("Oakland",row["team"]) or re.search("San Francisco",row["team"]):
            row["Metropolitan area"] = "San Francisco Bay Area"
        if re.search("St. Louis",row["team"]):
            row["Metropolitan area"] = "St. Louis"
        if re.search("Vegas",row["team"]):
            row["Metropolitan area"] = "Las Vegas"
        if re.search("Texas Rangers",row["team"]):
            row["Metropolitan area"] = "Dallas–Fort Worth"
        if re.search("Chicago",row["team"]):
            row["Metropolitan area"] = "Chicago"
        if re.search("Kansas City",row["team"]):
            row["Metropolitan area"] = "Kansas City"
        if re.search("San Diego",row["team"]):
            row["Metropolitan area"] = "San Diego"
        if re.search("Boston",row["team"]):
            row["Metropolitan area"] = "Boston"
        return row
    mlb_df = mlb_df.apply(city_associate,axis = 1)
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Dallas","Dallas–Fort Worth")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Carolina","Raleigh")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Colorado","Denver")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Arizona","Phoenix")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Florida","Miami–Fort Lauderdale")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Miami","Miami–Fort Lauderdale")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Minnesota","Minneapolis–Saint Paul")
    mlb_df["Metropolitan area"] = mlb_df["Metropolitan area"].replace("Washington","Washington, D.C.")
    mlb_df = mlb_df[mlb_df["year"] == 2018]

    ratioTable = pd.merge(cities,mlb_df,how = "inner",on = "Metropolitan area")
    ratioTable = ratioTable[["Metropolitan area","Population (2016 est.)[8]","W","L"]]
    def calc_wl_ratio(group):
            ratio = np.sum(group["W"].astype(float))/(np.sum(group["W"].astype(float)) + np.sum(group["L"].astype(float)))
            num = group.shape[0]
            return pd.Series({
                "Metro" : group["Metropolitan area"].iloc[0],
                "wl_ratio" : ratio,
                "population" : group["Population (2016 est.)[8]"].iloc[0],
                "num" : num
            })
    ratioTable = ratioTable.groupby("Metropolitan area").apply(calc_wl_ratio).reset_index(drop = True)

    ratioTable = ratioTable.set_index("Metro")
    

    return ratioTable

```

```python
def sports_team_performance():
    global mlb_df
    global nhl_df
    global nba_df
    global nfl_df
    # YOUR CODE HERE
    cities.replace("^[\s*—*\ *]$",np.nan,regex = True,inplace = True)
    cities.replace("\[.*\]","",regex = True,inplace = True)
    cities.replace("^$",np.nan,regex = True,inplace = True)
    cities["Metropolitan area"] = cities["Metropolitan area"].str.strip()
    cities.iloc[13,2] = np.nan
    nhl_df = nhl_df[nhl_df["GP"].str.isnumeric()]

    ratioTable_nba = nba()["wl_ratio"]

    ratioTable_mlb = mlb()["wl_ratio"]

    ratioTable_nfl = nfl()["wl_ratio"]

    ratioTable_nhl = nhl()["wl_ratio"]

    rel = [[np.nan for i in range(4)] for j in range(4)]

    temp = pd.merge(ratioTable_nfl,ratioTable_nba,left_index = True,right_index = True)
    rel[0][1] = stats.ttest_rel(temp["wl_ratio_x"],temp["wl_ratio_y"]).pvalue

    temp = pd.merge(ratioTable_nfl,ratioTable_nhl,left_index = True,right_index = True)
    rel[0][2] = stats.ttest_rel(temp["wl_ratio_x"],temp["wl_ratio_y"]).pvalue

    temp = pd.merge(ratioTable_nfl,ratioTable_mlb,left_index = True,right_index = True)
    rel[0][3] = stats.ttest_rel(temp["wl_ratio_x"],temp["wl_ratio_y"]).pvalue

    temp = pd.merge(ratioTable_nba,ratioTable_mlb,left_index = True,right_index = True)
    rel[1][3] = stats.ttest_rel(temp["wl_ratio_x"],temp["wl_ratio_y"]).pvalue

    temp = pd.merge(ratioTable_nba,ratioTable_nhl,left_index = True,right_index = True)
    rel[1][2] = stats.ttest_rel(temp["wl_ratio_x"],temp["wl_ratio_y"]).pvalue

    temp = pd.merge(ratioTable_nhl,ratioTable_mlb,left_index = True,right_index = True)
    rel[2][3] = stats.ttest_rel(temp["wl_ratio_x"],temp["wl_ratio_y"]).pvalue

    rel[1][0] = rel[0][1]
    rel[3][0] = rel[0][3]
    rel[3][1] = rel[1][3]
    rel[3][2] = rel[2][3]
    rel[2][1] = rel[1][2]
    rel[2][0] = rel[0][2]
    
    # Note: p_values is a full dataframe, so df.loc["NFL","NBA"] should be the same as df.loc["NBA","NFL"] and
    # df.loc["NFL","NFL"] should return np.nan
    sports = ['NFL', 'NBA', 'NHL', 'MLB']
    p_values = pd.DataFrame(rel, index=sports,columns = sports)
    
    assert abs(p_values.loc["NBA", "NHL"] - 0.02) <= 1e-2, "The NBA-NHL p-value should be around 0.02"
    assert abs(p_values.loc["MLB", "NFL"] - 0.80) <= 1e-2, "The MLB-NFL p-value should be around 0.80"
    return p_values
```
